```python
import pandas as pd
```

```python
data = pd.read_csv("/content/Pokemon.csv")
```

```python
data.head()
```

```python
hp_mean = data["HP"].mean()
def set_hp(val):
    if val < hp_mean:
        return "Low HP"
    elif val == hp_mean:
        return "neutral HP"
    else:
        return "High HP"

data['HP High or Low'] = data["HP"].apply(set_hp)

speed_mean = data["Speed"].mean()
def set_speed(val):
    if val < speed_mean:
        return "Low Speed"
    elif val == speed_mean:
        return "neutral Speed"
    else:
        return "High Speed"

data['Speed High or Low'] = data["Speed"].apply(set_speed)

data
```
