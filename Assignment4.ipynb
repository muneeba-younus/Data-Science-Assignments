# **Question 1:**

```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/netflix_titles.csv")

def split_comma(data):
  data = str(data)
  if data == "nan":
    return 0
  return len(data.split(","))

data['num_of_cast'] = data["cast"].apply(split_comma)
data
```
